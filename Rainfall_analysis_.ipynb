```python
import pandas as pd
import numpy as np

# Generate sample data for past rainfall
years = range(2000, 2021)
rainfall = np.random.randint(50, 200, size=len(years))  # Random rainfall values between 50mm and 200mm

# Create a DataFrame
rainfall_data = pd.DataFrame({'Year': years, 'Rainfall (mm)': rainfall})

# Save the DataFrame to a CSV file
rainfall_data.to_csv('past_rainfall_data.csv', index=False)

# Display the first few rows of the DataFrame
print(rainfall_data.head())

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load past rainfall data
past_rainfall_data = pd.read_csv('past_rainfall_data.csv')

# Analyze present rainfall
present_rainfall = 50  # Example present rainfall data (in mm)
average_past_rainfall = past_rainfall_data['Rainfall (mm)'].mean()
if present_rainfall > average_past_rainfall:
    print("Present rainfall is above average.")
elif present_rainfall < average_past_rainfall:
    print("Present rainfall is below average.")
else:
    print("Present rainfall is average.")

# Visualize past rainfall data
plt.figure(figsize=(10, 6))
plt.plot(past_rainfall_data['Year'], past_rainfall_data['Rainfall (mm)'], marker='o', linestyle='-')
plt.title('Past Rainfall Data in Chennai')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()

# Perform linear regression for future rainfall prediction
X = past_rainfall_data['Year'].values.reshape(-1, 1)
y = past_rainfall_data['Rainfall (mm)'].values.reshape(-1, 1)

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict future rainfall
future_years = np.array([2025, 2026, 2027]).reshape(-1, 1)  # Example future years
predicted_rainfall = model.predict(future_years)

# Visualize future rainfall prediction
plt.figure(figsize=(10, 6))
plt.plot(past_rainfall_data['Year'], past_rainfall_data['Rainfall (mm)'], marker='o', linestyle='-', label='Past Rainfall')
plt.plot(future_years, predicted_rainfall, marker='o', linestyle='--', color='red', label='Predicted Rainfall')
plt.title('Future Rainfall Prediction in Chennai')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.legend()
plt.show()
```
